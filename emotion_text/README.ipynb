### Описание проекта    
Анализ эмоционального состояния»

### Проблема
Эмоциональное состояние пациента может иметь важное значение для диагностики и планирования лечения. Однако пациентам может быть трудно выразить свои эмоции или быть осознанными о своем эмоциональном состоянии.

### Задача
Разработать концепцию и черновик продукта, который использует предобученные модели обработки естественного языка и анализа эмоций для определения эмоционального состояния пациента на основе его речи.

Студентам необходимо придумать решение, которое будет использовать предобученные технологии машинного обучения для определения эмоционального состояния пациента.

### Решение
Обучаем модель логистической регрессии для классификации на 6 классов эмоций.

### Какие данные использованы?
В работе используется датасет на 417000 строк:
https://huggingface.co/datasets/dair-ai/emotion/viewer/unsplit
